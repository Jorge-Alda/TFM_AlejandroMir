# Axion and Axion Like Particles (ALPS)

The Standard Model (SM) of particle physics is an exceptionally successful framework that describes electromagnetic, weak, and strong interactions with high precision. However, it leaves several deep theoretical questions unanswered. Among these, one of the most persistent and puzzling is the strong CP problem: the question of why quantum chromodynamics (QCD), despite allowing a CP-violating term in the Lagrangian, does not appear to violate CP symmetry in experiments.

## The Strong CP Problem

The QCD Lagrangian includes a topological term of the form:
$$\mathcal{L_\theta}=\theta\frac{g^2_s}{32\pi^2}G^a_{\mu\nu}\tilde{G}^{a\mu\nu}$$
where $G^a_{\mu\nu}$ is the gluon field tensor and $\tilde{G}^{a\mu\nu}=\frac{1}{2}\epsilon^{\mu\nu\rho\sigma}G^a_{\rho\sigma}$ is its dual. The parameter $\theta$ is dimensionless and, in principle, can take any value from 0 to $2\pi$. However, this term violates CP symmetry and contributes to the electric dipole moment (EDM) of the neutron:
$$d_n\approx 2.4\times 10^{-16}\theta e\cdot cm$$

Experimental bounds constrain the neutron EDM to be:
$$|d_n| < 1.8 \times 10^{-26} \, e \cdot \text{cm} \quad \text{(90\% CL, nEDM Collaboration, 2020)}$$
This implies an extremely fine-tuned bound on the parameter:
$$|\theta|\lesssim 10^{-10}$$
This unnatural smallness is what constitutes the strong CP problem.

## The Peccei–Quinn Mechanism and the Axion


In 1977, Peccei and Quinn proposed a dynamical solution to the strong CP problem by promoting $\theta$ to a dynamical field—effectively replacing the static parameter by the vacuum expectation value of a pseudo-Nambu–Goldstone boson
a(x), known as the axion. This field arises when a global anomalous $U(1)_PQ$ symmetry is spontaneously broken.
The axion couples to gluons as:
$$\mathcal{L_a}=\frac{a(x)}{f_a}\frac{g^2_s}{32\pi^2}G^a_{\mu\nu}\tilde{G}^{a\mu\nu}$$
This interaction effectively relaxes the CP-violating term dynamically: non-perturbative QCD effects generate a potential for the axion that minimizes it at $⟨a/f_a⟩+\theta=0$, solving the strong CP problem naturally.


## QCD Non-Perturbative Effects and the Axion Potential

The reason the axion dynamically cancels the $\theta$-angle lies in the non-perturbative structure of QCD. The strong interactions admit topologically nontrivial configurations known as instantons, which interpolate between degenerate classical vacua labeled by an integer winding number Q. These configurations generate a physically observable term:
$$Q=\frac{1}{32\pi^2}\int d^4x G^a_{\mu\nu}\tilde{G}^{a\mu\nu}$$
In the QCD path integral, this leads to a dependence on $\theta$ via the weight:
$$\mathcal{Z}(\theta)=\sum_Q\int [\mathcal{D}A]_Q e^{-S_{YM}+iQ\theta}$$
When the axion is introduced through the anomalous Peccei–Quinn symmetry, it enters the action as a dynamical field via the replacement $\theta\rightarrow\theta+a(x)/f_a$.
This results in a potential energy for the axion field generated by instanton effects. At leading order in a dilute instanton gas approximation, or equivalently using large-$N_c$ arguments (Witten–Veneziano), the effective potential takes the form:
$$V(a)=\chi_{QCD}\left(1-cos\left(\frac{a}{f_a}+\theta\right)\right)$$
Here, $\chi_{QCD}$ is the topological susceptibility of QCD, a nonperturbative quantity estimated to be $\chi\thicksim\Lambda^4_{QCD}$. This potential is periodic, reflecting the fact that the PQ symmetry is anomalous and explicitly broken by instanton effects.

The crucial result is that this potential is minimized when:
$$\frac{a}{f_a}+\theta\Big|_{min}=0\hspace{0.5cm}\Rightarrow\hspace{0.5cm} \langle a\rangle=-f_a\theta$$
Thus, the physical CP-violating term $\theta_{eff}=\theta+\frac{a}{f_a}$ is dynamically driven to zero. This mechanism solves the strong CP problem without fine-tuning.

Furthermore, the curvature of the potential around its minimum determines the axion mass:
$$m^2_a=\frac{\partial^2V}{\partial a^2}\Big|_{a=-f_a\theta}=\frac{\chi_{QCD}}{f_a^2}$$

## From QCD Axions to Axion-Like Particles (ALPs)


Although the original QCD axion was predicted to have properties dependent on the symmetry breaking scale $f_a$, experimental constraints have pushed $f_a$ to extremely high values ($f_a\gtrsim 10^9GeV$), rendering the axion very light and very weakly coupled.

This realization inspired the theoretical consideration of more general pseudoscalar particles—Axion-Like Particles (ALPs)—which:

- Do not necessarily solve the strong CP problem,

- May have arbitrary mass and coupling scales,

- Still arise as pseudo-Nambu–Goldstone bosons in many extensions of the SM, especially in string theory compactifications.

These ALPs can couple to photons, fermions, or gauge bosons via effective dimension-5 operators:
$$\mathcal{L}_{ALP}\supset\frac{1}{4}g_{a\gamma}aF_{\mu\nu}\tilde{F}^{\mu\nu}+ ...$$

## The QCD Axion

The QCD axion is a pseudoscalar particle emerging from the spontaneous breaking of a global anomalous symmetry. Its dynamics provide a natural solution to the strong CP problem by relaxing the effective CP-violating angle $\theta_{eff}=\theta+\frac{a}{f_a}$ to zero. While the idea is general, different UV completions of the Peccei–Quinn mechanism result in different phenomenological axion models. The two most prominent are:

- The KSVZ model (Kim–Shifman–Vainshtein–Zakharov)

- The DFSZ model (Dine–Fischler–Srednicki–Zhitnitsky)

### Axion mass and couplings

Due to its coupling to gluons, the axion acquires a potential from QCD non-perturbative effects, leading to a calculable mass. From chiral perturbation theory (ChPT), the mass of the axion in the QCD vacuum is given by:
$$m_a^2 = \frac{m_u m_d}{(m_u + m_d)^2} \cdot \frac{m_\pi^2 f_\pi^2}{f_a^2} \Rightarrow m_a \approx 5.7 \, \mu\text{eV} \left( \frac{10^{12} \, \text{GeV}}{f_a} \right)
$$
Similarly, the axion–photon coupling is model-dependent but generally of the form:
$$\mathcal{L}_{a\gamma} = -\frac{1}{4} g_{a\gamma} \, a F_{\mu\nu} \tilde{F}^{\mu\nu} \hspace{0.5cm} \text{with} \hspace{0.5cm} g_{a\gamma} = \frac{\alpha}{2\pi f_a} \left( \frac{E}{N} - \frac{2}{3} \frac{4 + z}{1 + z} \right)$$
where:
- $z=m_u/m_d\approx 0.56$
- E/N is the model-dependent ratio of electromagnetic to color anomalies.

### The KSVZ Model

The KSVZ axion model introduces a new heavy vector-like quark Q, charged under QCD but neutral under the SM electroweak group. It couples to a complex scalar $\phi$ which carries the PQ charge:
$$\mathcal{L}_{\text{KSVZ}} \supset y_Q \, \phi \bar{Q}_L Q_R + \text{h.c.}
$$
- The PQ symmetry is broken when $\phi$ acquires a VEV $f_a$
- The axion arises as the phase of $\phi$
- The SM fermions do not carry PQ charges, so the axion does not couple to fermions at tree level.
- The coupling to photons is purely via heavy quark loops.

Typical anomaly ratio: E/N=0, leading to a relatively small photon coupling.


### The DFSZ Model

The DFSZ axion model extends the SM scalar sector with two Higgs doublets $H_u$, $H_d$ and a complex scalar singlet $\phi$, all charged under $U(1)_{PQ}$. The Yukawa couplings involve:

$$\mathcal{L}_{\text{DFSZ}} \supset y_u \, \bar{Q}_L H_u u_R + y_d \, \bar{Q}_L H_d d_R + \lambda \, \phi H_u^\dagger H_d + \text{h.c.}
$$

- The PQ symmetry is broken when $\phi$ gets a VEV $f_a$, giving rise to the axion.
- Unlike KSVZ, the axion couples directly to SM fermions.
- The electromagnetic anomaly is nonzero: E/N=8/3, leading to a larger photon coupling.


Although the original Yukawa Lagrangian does not show an explicit coupling between the axion $a$ and the fermions, this interaction becomes apparent once one performs a field redefinition to remove the axion from the scalar sector. This procedure is standard in non-linear realizations of spontaneous symmetry breaking.
In particular, PQ-charged fermions are rephased to absorb the axion's dependence:
$$\psi\rightarrow e^{iX_\psi a(x)/f_a}\psi$$
This transformation introduces derivative couplings of the axion to fermions in the kinetic terms, which at low energies give rise to pseudo-scalar interactions:

$$\mathcal{L}_{a\psi}=\frac{C_\psi m_\psi}{f_a}a(x)\bar{\psi}i\gamma^5\psi$$
where $C_\psi$ depends on the PQ charge assignment of each fermion. In the DFSZ model, since SM fermions carry PQ charges, these couplings exist at tree level.

This is in contrast to the KSVZ model, where SM fermions are PQ-neutral, and axion–fermion couplings arise only via quantum loops.




### Experimental implications

- The axion mass and coupling are inversely proportional to the PQ scale $f_a$, which must be $\gtrsim 10^9 GeV$ from astrophysical bounds.

- QCD axions are viable cold dark matter candidates if produced via the misalignment mechanism.

- The two benchmark models define a "QCD axion band" in the ($m_a$, $g_{a\gamma}$) plane that is used to guide experimental searches.



## Axion-Like Particles (ALPs)

Axion-like particles (ALPs) are a class of pseudoscalar bosons that generalize the properties of the QCD axion. While the QCD axion has a well-defined relation between its mass $m_a$ and decay constant $f_a$ due to its coupling to gluons and the structure of QCD, ALPs are characterized by a *decoupling* of these two parameters. That is, an ALP can have:

- Arbitrary mass $m_a$
- Arbitrary couplings $g_{a\gamma}, g_{aff}, g_{agg}$
- Not necessarily solve the strong CP problem

They are typically described as pseudo–Nambu–Goldstone bosons of a spontaneously broken approximate global symmetry $U(1)_{\text{ALP}}$, which is not necessarily anomalous under QCD, though it may be anomalous under electromagnetism or other gauge symmetries.

---

### General Effective Lagrangian

At energies below the symmetry breaking scale $f_a$, the leading-order couplings of an ALP $a(x)$ are described by dimension-5 effective operators:

$$
\begin{aligned}
\mathcal{L}_{\text{ALP}} \supset \, &
\frac{1}{2} (\partial_\mu a)^2 - \frac{1}{2} m_a^2 a^2 +
\frac{g_{a\gamma}}{4} a F_{\mu\nu} \tilde{F}^{\mu\nu} \\
& + \frac{g_{agg}}{4} a G_{\mu\nu} \tilde{G}^{\mu\nu}
+ \sum_f \frac{C_f}{2 f_a} \partial_\mu a \, \bar{\psi}_f \gamma^\mu \gamma^5 \psi_f
\end{aligned}
$$

In contrast to the QCD axion, where:

$$
m_a \sim \frac{\Lambda_{\text{QCD}}^2}{f_a}, \quad g_{a\gamma} \sim \frac{\alpha}{2\pi f_a}
$$

ALPs can have, e.g., $m_a \sim \text{eV}$ and $f_a \sim 10^9\,\text{GeV}$, or even ultralight $m_a \sim 10^{-22} \, \text{eV}$ with extremely weak couplings—depending on the UV completion.

---

### Origin in Theoretical Frameworks

ALPs arise generically in many extensions of the Standard Model:

- **String theory** compactifications produce large families of ALPs (the "axiverse")
- **Composite Higgs** and **extra-dimensional models** often predict pseudo-Goldstone bosons
- **Relaxion-type** models utilize ALP-like fields for solving the hierarchy problem


### Astrophysical and Cosmological Significance

ALPs, like the QCD axion, are excellent candidates for cold dark matter and also affect several astrophysical processes:

- **Energy loss** in stars and white dwarfs via $\gamma \rightarrow a$
- **CMB distortions** via photon–ALP mixing
- **Fuzzy dark matter** scenarios for $m_a \sim 10^{-22} \, \text{eV}$
- ALP fields may play a role in inflation, baryogenesis, or moduli stabilization

### Summary: Differences Between QCD Axions and ALPs

| Feature                 | QCD Axion                 | ALP (general)               |
|-------------------------|---------------------------|-----------------------------|
| Solves strong CP?       | Yes                       | Not necessarily             |
| Mass–coupling relation  | $m_a \sim 1/f_a$      | Arbitrary                   |
| Coupling to gluons      | Required                  | Optional                    |
| Common origin           | PQ symmetry               | String theory, BSM, etc.    |
| Dark matter candidate?  | Yes                       | Yes (broad possibilities)   |


## Experimental Constraints and Detection of Axions and ALPs

Axions and axion-like particles (ALPs) can be searched for across a wide range of experiments and astrophysical observations, exploiting their characteristic couplings to photons, nucleons, and electrons. These searches span several decades of mass and coupling, and fall into the following categories:

---

### 1. Stellar and Supernova Constraints

ALPs can be produced inside stars via their coupling to photons or nucleons and escape freely, providing an additional energy loss channel. This leads to observable consequences in stellar evolution:

- **Helium-burning (HB) stars and red giants** constrain the ALP–photon coupling:
  
  $$
  g_{a\gamma} \lesssim 6.6 \times 10^{-11}~\text{GeV}^{-1}, \quad \text{for } m_a \lesssim 10~\text{keV}
  $$

- **Supernova SN1987A** places strong bounds on ALPs coupled to nucleons and photons by requiring the observed neutrino burst to remain unaffected:

  $$
  g_{a\gamma} \lesssim 5 \times 10^{-12}~\text{GeV}^{-1}, \quad \text{for } m_a \lesssim 10^{-9}~\text{eV}
  $$

📄 [2104.05777](https://arxiv.org/abs/2104.05777) – Bounds from HB stars and SN1987A

---

### 2. Helioscopes: CAST and IAXO

Helioscopes aim to detect solar axions through their conversion into X-ray photons in a magnetic field aligned with the solar direction.

- **CAST (CERN Axion Solar Telescope)** sets leading limits for $m_a \lesssim 1 \, \text{eV}$:

  $$
  g_{a\gamma} \lesssim 6.6 \times 10^{-11}~\text{GeV}^{-1}
  $$

- **IAXO (International Axion Observatory)**, the next-generation helioscope, will improve sensitivity by over an order of magnitude and explore realistic axion models (KSVZ/DFSZ).

📄 [2006.11894](https://arxiv.org/abs/2006.11894) – IAXO physics case

---

### 3. Haloscopes: Dark Matter Axion Detection

Haloscopes search for cold axion dark matter in the Milky Way halo via its resonant conversion into photons in a microwave cavity immersed in a magnetic field:

- **ADMX** targets axions in the “classic” mass range:

  $$
  m_a \sim \mu\text{eV}, \quad f_a \sim 10^{12}~\text{GeV}
  $$

- **HAYSTAC**, **DMRadio**, and **QUAX** explore other mass windows using tunable resonators and broadband detection.

📄 [2105.05850](https://arxiv.org/abs/2105.05850) – Status of axion haloscope searches

---

### 4. Light-Shining-Through-Walls (LSW)

In LSW experiments, a laser beam propagates in a strong magnetic field, producing ALPs via $\gamma \to a$. These ALPs traverse an optical barrier and reconvert into photons on the other side.

- **ALPS II (DESY)** will improve sensitivity to ALPs with:

  $$
  g_{a\gamma} \sim 2 \times 10^{-11}~\text{GeV}^{-1}, \quad m_a \lesssim \text{meV}
  $$

📄 [2206.13549](https://arxiv.org/abs/2206.13549) – ALPS II technical design and status

---

### 5. Exclusion Limits Summary

The combined results from astrophysical bounds, helioscopes, haloscopes, and laboratory searches constrain a large portion of the $(m_a, g_{a\gamma})$ parameter space.

Typical QCD axion models lie along a band defined by:

$$
g_{a\gamma} \propto \frac{1}{f_a}, \quad m_a \propto \frac{1}{f_a}
$$
